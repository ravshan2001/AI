```python
import tensorflow as tf
import keras
import numpy 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
plt.figure(figsize=(1,1))
plt.imshow(x_train[0])
```

```python
x_train.shape
```

```python
def plot_sample(index):
    plt.figure(figsize=(1,1))
    plt.imshow(x_train[index])
```

```python
plot_sample(len(x_train)-1)
```

```python
plot_sample(2)
```

```python
classes = ['airplane', 'automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
classes[y_train[2][0]]
```

```python
x_train_scaled = x_train/255
x_test_scaled = x_test/255
```

```python
x_train_scaled[0]
```

```python
y_train_categorical = keras.utils.to_categorical(
        y_train, num_classes=10, dtype='float32'
)
```

```python
y_train_categorical[0:10]
```

```python
y_test_categorical = keras.utils.to_categorical(
        y_test, num_classes=10, dtype='float32'
)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(3000,activation='relu'),
    keras.layers.Dense(1000,activation='relu'),
    keras.layers.Dense(500,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(optimizer='SGD',
             loss = 'categorical_crossentropy',
             metrics='accuracy')
```

```python
model.fit(x_train_scaled,y_train_categorical, epochs=30)
```

```python
model.fit(x_train_scaled,y_train_categorical, epochs=20)
```

```python
classes[np.argmax(model.predict(x_test_scaled[400]))]
```

```python
plt.figure(figsize=(1,1))
plt.imshow(x_test_scaled[400])
```
